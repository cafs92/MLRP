# Simple Percetron
Perceptron Simples aplicado à base de dados Iris Flower Classification.

Um perceptron simples é um classificador binário simples, que mapeia uma entrada x a uma saída y, onde y é dado por:


```python
import numpy as np
from sklearn.model_selection import train_test_split
from src import utils as ut
```

As funções a seguir são necessárias para o funcionamento do algoritmo, mas não necessariamente fazem parte
do mesmo. 

->insertbias insere o limiar (-1) em toda a base de dados.


```python
def insertbias(dataset):

    new = []

    for i in range(len(dataset)):

        new.append(np.insert(dataset[i], 0, -1))

    return np.asarray(new)
```

Treinamento:

O treinamento de um perceptron simples se dá atualizando os pesos w das entradas e do bias.
Isto é feito através de um somatório ponderado observando-se o erro e no fim
de cada execução.
##As fórmulas iriam aqui se eu soubesse colocá-las##

```python
def train(ts, yts ,w, c = 300, eta = 0.1):

    counter = 0

    C =0

    while counter < c:

        for i, j in zip(ts, yts):

            u = np.dot(w, i)

            y = np.where(u >= 0.0, 1, -1)

            e = i[len(i) - 1] - y

            w = w + eta*e*j

        counter += 1

        np.random.shuffle(ts)

    for i, j in zip(ts, yts):

        u = np.dot(w, i)

        y = np.where(u >= 0.0, 1, -1)

        e = j - y

        if e != 0:

            C += 1



    print('Treino: ', 1 - (C/len(ts)))

    return w
```

Teste

```python
def test(w, vs, yvs):

    c = 0

    for i, j in zip(vs, yvs):

        u = np.dot(w, i)

        y = np.where(u >= 0.0, 1, -1)

        e = j - y

        if e != 0:

            c+=1



    return 1 - (c/len(vs))
```

Função principal

Aqui vemos o funcionamento estritamente estruturado do algoritmo.



```python
def simple_perc():
    c = 100
    x, y = ut.readdata()
    w = np.zeros(x.shape[1]+1)

    X = insertbias(x)

    x_train, x_test, y_train, y_test = train_test_split(X, y, test_size=0.2)#Holdout 80/20
    

    w = train(x_train, y_train, w, c)

    error = test(w, x_test, y_test)

    print('Teste:', error)

    return w

for i in range(10):
    simple_perc()
    print('Oi')
```
